```python
plt.subplots(figsize=(25,20))
correlation_mat = df.corr()
sns.heatmap(correlation_mat, annot = True)
plt.title("Correlation matrix of covid data of states")

plt.xlabel("states")

plt.ylabel("states")
plt.show()
```
